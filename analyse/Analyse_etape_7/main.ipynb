```python
import csv
import matplotlib.pyplot as plt
import numpy as np

def mse(actual, pred): 
    actual, pred = np.array(actual), np.array(pred)
    return np.square(np.subtract(actual,pred)).mean()
```

```python
u_th = []
v_th = []
p_th = []
x_th = []
y_th = []

i = 0
with open('u_velocity_y=0.5.csv') as file:
    reader = csv.reader(file, delimiter=',')
    for line in reader:
        # print(line)
        if (i!=0):
            y_th.append(float(line[1]))
            u_th.append(float(line[2]))
        i+= 1


plt.plot(y_th, u_th, label = r'Thorique')

u = []
v = []
p = []
x = []
y = []

i = 0
with open('CD4_x=0.5.csv') as file:
    reader = csv.reader(file, delimiter=',')
    for line in reader:
        # print(line)
        if (i!=0):
            x.append(float(line[6]))
            y.append(float(line[7]))
            u.append(float(line[2]))
            v.append(float(line[3]))
            p.append(float(line[1]))
        i+= 1

plt.plot(y, u, label = r'Rsultat CD4')
plt.text(0.4, 0.8, 'mean square error CD4= %5.3e' % mse(u, u_th[::-1]) )

u = []
v = []
p = []
x = []
y = []

i = 0
with open('CD2_x=0.5.csv') as file:
    reader = csv.reader(file, delimiter=',')
    for line in reader:
        # print(line)
        if (i!=0):
            x.append(float(line[6]))
            y.append(float(line[7]))
            u.append(float(line[2]))
            v.append(float(line[3]))
            p.append(float(line[1]))
        i+= 1

plt.plot(y, u, label = r'Rsultat CD2')
plt.text(0.4, 0.9, 'mean square error CD2= %5.3e' % mse(u, u_th[::-1]) )

u = []
v = []
p = []
x = []
y = []

i = 0
with open('UR1_x=0.5.csv') as file:
    reader = csv.reader(file, delimiter=',')
    for line in reader:
        # print(line)
        if (i!=0):
            x.append(float(line[6]))
            y.append(float(line[7]))
            u.append(float(line[2]))
            v.append(float(line[3]))
            p.append(float(line[1]))
        i+= 1

plt.plot(y, u, label = r'Rsultat UR1')
plt.text(0.4, 1, 'mean square error UR1= %5.3e' % mse(u, u_th[::-1]) )

plt.xlabel(r'$y$ $[m]$')
plt.ylabel(r'$u$ $[m.s^{-1}]$')
plt.legend()

plt.savefig('u_x=0.5.png', dpi=300)
plt.show()
```
