```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
initial_solution_temp = np.loadtxt('../output/initial_solution.dat')
final_solution_temp = np.loadtxt('../output/final_solution.dat')

Nx = 201
space_grid = np.zeros(Nx)
final_solution = np.zeros(Nx)
initial_solution = np.zeros(Nx)

for i in range(Nx):
    space_grid[i] = final_solution_temp[i][0]
    initial_solution[i] = initial_solution_temp[i][1]
    final_solution[i] = final_solution_temp[i][1]

x_0 = 0.2
delta = 0.05
c = 1.0
t_f = 0.5
theorical_solution = [np.exp(-((space_grid[i]-(x_0 + c*t_f))/delta)**2) for i in range(len(space_grid))]


plt.clf()
plt.plot(space_grid, initial_solution, label='t = 0')
plt.plot(space_grid, final_solution, label='résultat à t = 0,5s')
plt.plot(space_grid, theorical_solution, label='théorique à t = 0,5s')
plt.xlabel('x [m]')
plt.ylabel(r'u [$m.s^{-1}$]')
plt.title('Resultat de l étape 1')
plt.legend()
plt.savefig('plot_etape1.png', dpi = 300)
plt.show()
```
