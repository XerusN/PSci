```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.clf()
num_colonne = 3

value_file = np.loadtxt('jacobi_short_5.dat')

dx = np.zeros(int(value_file.size/num_colonne))
mean_iter_number = np.zeros(int(value_file.size/num_colonne))
computation_time = np.zeros(int(value_file.size/num_colonne))

for i in range(int(value_file.size/num_colonne)):
    dx[i] = value_file[i][0]
    mean_iter_number[i] = value_file[i][1]
    computation_time[i] = value_file[i][2]

plt.plot(dx, computation_time, label='Jacobi')

############

value_file = np.loadtxt('gauss-seidel_short_5.dat')

dx = np.zeros(int(value_file.size/num_colonne))
mean_iter_number = np.zeros(int(value_file.size/num_colonne))
computation_time = np.zeros(int(value_file.size/num_colonne))

for i in range(int(value_file.size/num_colonne)):
    dx[i] = value_file[i][0]
    mean_iter_number[i] = value_file[i][1]
    computation_time[i] = value_file[i][2]

plt.plot(dx, computation_time, label='Gauss-Seidel')

############

value_file = np.loadtxt('relaxation_short_5.dat')

dx = np.zeros(int(value_file.size/num_colonne))
mean_iter_number = np.zeros(int(value_file.size/num_colonne))
computation_time = np.zeros(int(value_file.size/num_colonne))

for i in range(int(value_file.size/num_colonne)):
    dx[i] = value_file[i][0]
    mean_iter_number[i] = value_file[i][1]
    computation_time[i] = value_file[i][2]

plt.plot(dx, computation_time, label='Surrelaxation')

############

value_file = np.loadtxt('conjugate_short_5.dat')

dx = np.zeros(int(value_file.size/num_colonne))
mean_iter_number = np.zeros(int(value_file.size/num_colonne))
computation_time = np.zeros(int(value_file.size/num_colonne))

for i in range(int(value_file.size/num_colonne)):
    dx[i] = value_file[i][0]
    mean_iter_number[i] = value_file[i][1]
    computation_time[i] = value_file[i][2]

plt.plot(dx, computation_time, label='Conjugate')
plt.xlabel()
plt.loglog()
plt.legend()
plt.savefig('linear_solving_time.png', dpi = 300)
plt.show()
```
